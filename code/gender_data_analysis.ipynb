### GET LAC from IEC-GPI 

```python
#Add column regional groups to df_gpi

def fx_extract_reg_group(cod):
    for _,item in df_countries_list.iterrows():    #_ VAR QUE NO USAS
        if cod == item['iso3']:
            return item['un_regional_groups_s']
    return ''

df_gpi['ONUGr']=df_gpi['iso3c'].apply(lambda x: fx_extract_reg_group(x))
```

```python
df_gpi.head()
```

```python
#Determine a dataframe with values from latin american countries only

df_gpi_LAC=df_gpi.loc[df_gpi['ONUGr']=='G-ALC',:]
```

```python
#Pivot table
df_CEPAL_stat_femicides=df_CEPAL_stat_femicides.pivot(index='country_sp', columns='year', values='femicides_number')
```

#### Add iso3 code

```python
#Add an iso3 column with corresponding values to df_CEPAL_stat_femicides from df_countries_list_LAC

def fx_extract_iso3(country):
    for _,item in df_countries_list_LAC.iterrows():    
        if country == item['country_sp']:
            return item['iso3']
    return ''

df_CEPAL_stat_femicides['iso3']=df_CEPAL_stat_femicides['country_sp'].apply(lambda x: fx_extract_iso3(x))
```
